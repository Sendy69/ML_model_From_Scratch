```python
import numpy as np
import pandas as pd
import tqdm
import matplotlib.pyplot as plt
from collections import Counter

# Calcul de la distance euclidienne entre 02 points
# Que se passe t'il lorsqu'il y'a plusieurs caractèristiques??

def euclidean_distance(point1, point2):
    distance = 0
    for a, b in zip(point1, point2):
        distance += (b-a)**2
    return distance ** 2 
    

class KNN:
    def __init__(self, k):
        self.k = k
    
    def fit(self, x_train, y_train):
        self.x_train = x_train
        self.y_train = y_train

    def predict(self, new_point):
        #compute the distances and 
        distance_couple = [euclidean_distance(new_point, each_x) for each_x in self.x_train]

        # sorted list of distances
            ### 
        distances_tuples = sorted(list(enumerate(distance_couple)), key= lambda x: x[1])[:self.k]
        #     ###
        # sorted_list_of_distances = sorted(list_of_distances)
        # #get the closest K 
        # neighbors = list(sorted_list_of_distances)
        
        indexes = [distance_tuple[0] for distance_tuple in distances_tuples]
        # get the classifications


        classifications = [self.y_train[index] for index in indexes]

        #classifications = [ self.y_train[index] for index , _ in enumerate(neighbors)]
        print(classifications)

        
        
        ## Get Max classification
       
        #max_class = max(classifications, key= classifications.count)
        max_class = Counter(classifications).most_common()
        print(max_class)
       

        return max_class[0][0]
    
    def test(self, x_tests, y_tests):
        y_tests_predicted = [self.predict(x_test) for x_test in tqdm.tqdm(x_tests)]
        good_prediction = 0
        for y_test_predicted, y_test in zip(y_tests_predicted, y_tests):  
            if y_test_predicted == y_test:
                good_prediction += 1
            return good_prediction / len(y_tests)



        
    #cross_validation from scratch
        #Grid search cv
```

```python

df = pd.read_csv("KNN - K Means/ushape.csv", header=None).sample(frac=1, random_state=2)


X = df[[0, 1]].values
Y = df[2].values

index_split = int(len(Y)* 0.7)

x_trains = X[: index_split]
x_tests = X[index_split: ]



y_trains = Y[: index_split]
y_tests = Y[index_split: ]
knn_model = KNN(k=5)
knn_model.fit(x_trains, y_trains)
print(knn_model.test(x_tests, y_tests))
```
